```python
# Filtrowanie danych z lambdą
bios["height"] = bios["height_cm"].apply(lambda x: "Short" if x < 165 else("Medium" if x < 180 else "Tall"))
bios.head(20)
```

Można też używać funkcji do filtracji danych

```python
def categorize_athlete(row):
    if row["height_cm"] >= 165:
        return "Lightweight"
    elif row["height_cm"] >= 180:
        return "Medium"
    else:
        return "Happy"
    
bios["height"] = bios.apply(categorize_athlete, axis=1)
# axis 1 to wiersze, axis 0 to kolumny
```

<big>Handling NULL Values<big>:

```python
coffe.loc[[0,1], "Units Sold"] = np.nan    # np.nan zamienia podane wartosci czyli wartosci pod indeksem 0 i 1 w kolumnie coffe type na Nan
coffe.isna().sum()                          # isna() sprawdza czy dana wartość ma typ Nan, sum() liczy ile ich jest
coffe.fillna(coffe["Units Sold"].mean())    # fillna() wypełnia wartości non określonom rzeczą w tym przpadku średnia ze sprzedaży
coffe["Units Sold"].interpolate() # interpolate() znajduję patern w zbiorze i wstawia wartość w puste miejsca (NaN) pasującą do niego.(Bierze sąsiednie wartości i widzi czy jest malejący, rosnący itp, oraz wpisuje pasującą wartość).  Nie działa na pierwszych wartosciach bo nie ma skąd zabrać sąsiednich wartosci
coffe["Units Sold"] = coffe["Units Sold"].interpolate()     # <- to jest to samo co linijkę wyżej tylko zapisze na stałe do dataframu coffe

coffe.dropna(subset=["Units Sold"])          # usuwa wszystkie Nan które dodaliśmy, subset to kolumny które ma to objąć(jak wszystkie to zostawić puste ()
coffe[coffe["Units Sold"].notna()]           # zwróci tablicę bez wartości Nan, samo coffe.notna() sprawdzi które wartości nie są Nan - odwrotność .isna()
```

<big>Aggregating data - zbieranie danych<big>

```python
bios["born_city"].value_counts()            # Zlicza ile razy wystąpiła dana wartość
bios[bios["born_country"] == "USA"]["born_region"].value_counts()   # [bios["born_country"] == "USA"] -- to wybiera tylko atletów z usa ["born_region"].value_counts() zlicza ile razy wystąpił dany region

coffe.groupby(["Coffee Type"])["Units Sold"].sum()      # grupuje po Coffe Type a nastepnie sumuje "Units Sold"
coffe.groupby(["Coffee Type", "Day"]).agg({"new price": 'sum', "Units Sold":"mean"})   # .agg pozwala na operacja na większej ilości kolumn, {"nazwa kolumny": "operacja"} 

pivot = coffe.pivot(columns="Coffee Type", index="Day", values="revenu")  # tworzy nową tabelę, columns to klumny, index to wiersze, a values to wartosci
pivot
```

```python
coffe["yesterday revenu"] = coffe["revenu"].shift(2)
# .shift(2) przesuwa wartości w wierszach o określoną liczbę, dwa miejsca w dół
```

```python
bios["height rank"] = bios["height"].rank(ascending=False)  # rank, nadaje wartosc na podstawie wzrostu, im wyższa tym ktoś wyższy 
bios
```

```python
coffe["cumulative income"] = coffe["revenu"].cumsum()           # sumuje po kolei wartosci z revenu
coffe
```

```python
coffe["3day_sum"] = coffe["revenu"].rolling(3).sum()        # bierze 3 wcześniejsze wartości "revenu" i je sumuje w nowej kolumnie
coffe
```
